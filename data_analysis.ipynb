```python
table = []

for i in range(17):
    sd = 'gamesd{}'.format(i)
#     print(sd)

    game1 = data[data[sd]==1]

    n_games = len(np.unique(np.array(game1['game'])))
    session = np.unique(np.array(game1['session']))
    treatment = np.unique(np.array(game1['treat']))
    batch_me = np.unique(np.array(game1['batch_me']))
    n_buyers = len(np.unique(np.array(game1[(game1['buyer'] == 1) & (game1['game'] == 1)]['id'])))
    n_sellers = len(np.unique(np.array(game1[(game1['buyer'] == 0) & (game1['game'] == 1)]['id'])))
    
    table.append([sd, n_games, session, treatment, batch_me, n_buyers, n_sellers])
                        
table = pd.DataFrame(table, columns = ['SD', '# games', 'Session', 'treatment', 'batch_me', '# Buyers', '# sellers'])
table
```
