# Look at offer behaviour 

```python
data = pd.read_stata('data/DataTimeSeries.dta')
treatment1 = data[(data['gamesd0']==1)].copy()
treatment1 = data.copy()
treatment1['game_uniq_id'] = (np.array(treatment1['id']) + np.array(treatment1['game']) * 1000)
```

```python
buyers = treatment1[treatment1['buyer'] == 1]
sellers = treatment1[treatment1['buyer'] == 0]
```

```python
init_offer_buyer = buyers['d'][buyers['fwi'] == 'FALSE']
plt.hist(init_offer_buyer);
```

```python
init_offer_seller = sellers['d'][sellers['fwi'] == 'FALSE']
plt.hist(init_offer_seller);
```

# Seller behvaiour

```python
import scipy.stats as st

args = st.expon.fit(init_offer_seller)

print(args)
print(st.kstest(init_offer_seller, st.expon.cdf, args))

plt.figure(figsize=(6,6))

h = plt.hist(init_offer_seller, label='Empirical data', bins=15);
xx = np.linspace(0., init_offer_seller.max(), 100)
yy = st.expon.pdf(xx, *args)
nbins = len(h[0])
omax = init_offer_seller.max()
plt.plot(xx,yy * len(init_offer_seller) * omax / nbins, lw=5, label='Exponential dist.');
plt.legend()

plt.savefig('figs/hist_seller.png', dpi=400)
```

```python
import scipy.stats as st

dist = st.norm
args = dist.fit(init_offer_seller)

print(args)
print(st.kstest(init_offer_seller, dist.cdf, args))

h = plt.hist(init_offer_seller);
xx = np.linspace(0., init_offer_seller.max(), 100)
yy = dist.pdf(xx, *args)
nbins = len(h[0])
omax = init_offer_seller.max()
plt.plot(xx,yy * len(init_offer_seller) * omax / nbins, lw=4);
```

# Buyer behaviour

```python
import scipy.stats as st

args = st.expon.fit(init_offer_buyer)

print(args)
print(st.kstest(init_offer_buyer, st.expon.cdf, args))

h = plt.hist(init_offer_buyer);
xx = np.linspace(0., init_offer_buyer.max(), 100)
yy = st.expon.pdf(xx, *args)
nbins = len(h[0])
omax = init_offer_buyer.max()
plt.plot(xx,yy * len(init_offer_buyer) * omax / nbins, lw=4);
```

```python
import scipy.stats as st

offer_scaled = init_offer_buyer / init_offer_buyer.max()
offer_scaled = offer_scaled / 0.99 + 0.005
args = st.beta.fit(offer_scaled)

print(args)
print(st.kstest(offer_scaled, st.beta.cdf, args))

nbins = 20
h = plt.hist(offer_scaled, bins=nbins);
xx = np.linspace(0., offer_scaled.max(), 100)
yy = st.beta.pdf(xx, *args)
nbins = len(h[0])
omax = offer_scaled.max()
plt.plot(xx,yy * len(offer_scaled) * omax / nbins);
```

```python
import scipy.stats as st
dist = st.halflogistic
args = dist.fit(init_offer_buyer)

print(args)
print(st.kstest(init_offer_buyer, dist.cdf, args))

plt.figure(figsize=(6,6))
h = plt.hist(init_offer_buyer, label='Empirical data');
xx = np.linspace(0., init_offer_buyer.max(), 100)
yy = dist.pdf(xx, *args)
nbins = len(h[0])
omax = init_offer_buyer.max()
plt.plot(xx,yy * len(init_offer_buyer) * omax / nbins, lw=5, label='Half-logistic');
plt.legend()

plt.savefig('figs/hist_buyer.png', dpi=400)
```

```python
import scipy.stats as st

args = st.halfnorm.fit(init_offer_buyer)

print(args)
print(st.kstest(init_offer_buyer, st.halfnorm.cdf, args))

h = plt.hist(init_offer_buyer);
xx = np.linspace(0., init_offer_buyer.max(), 100)
yy = st.halfnorm.pdf(xx, *args)
nbins = len(h[0])
omax = init_offer_buyer.max()
plt.plot(xx,yy * len(init_offer_buyer) * omax / nbins, lw=4);
```
